# Preparing the data

```python
from sklearn import preprocessing

# separate the data from the target attributes
Y = mushroom_df['class']
X = mushroom_df.drop('class',axis=1)
# normalize the data attributes
normalized_X = preprocessing.normalize(X)
```

```python
x=pd.DataFrame(normalized_X,columns=X.columns)
```

# Analysing the accuracy per model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report,roc_auc_score,roc_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
best_rstate=0
accu=0
for i in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(normalized_X,Y,test_size=0.25,random_state=i)
    mod=LogisticRegression()
    mod.fit(x_train,y_train)
    y_pred=mod.predict(x_test)
    tempaccu=accuracy_score(y_test,y_pred)
    if tempaccu>accu:
        accu=tempaccu
        best_rstate=i

print(f"Best Accuracy {accu*100} found at random state {best_rstate} ")
```

# Hypertuning the RandomForestClassifier

```python
x_train,x_test,y_train,y_test=train_test_split(normalized_X,Y,test_size=0.20,random_state=42)
Best_Model = RandomForestClassifier()
parameters = {"n_estimators":[10,20,30,100],"criterion":['gini', 'entropy'],"max_features":['auto', 'sqrt', 'log2']}
gs=GridSearchCV(Best_Model,parameters)
gs.fit(x_train,y_train)
gs.best_estimator_.get_params()
```

We get only n_estimator as 10 which is not default, rest all the values are default.

# Finalising the model

```python
x_train,x_test,y_train,y_test=train_test_split(normalized_X,Y,test_size=0.20,random_state=42)
mod=RandomForestClassifier(n_estimators=10)
mod.fit(x_train,y_train)
y_pred=mod.predict(x_test)
tempaccu=accuracy_score(y_test,y_pred)
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(mod,X,Y,cv=30,scoring='accuracy').mean()
print("std: {}".format(score.std()))
print("cross_val_score: {}".format(score*100))
```

Our Model has a good cross validation score and will perform well
